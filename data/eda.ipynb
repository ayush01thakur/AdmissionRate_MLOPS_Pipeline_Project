```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
data= pd.read_csv("university_admission.csv")
data.head()
```

```python
data.info()
```

let's check for the correlations to find the important features only for the training purpose

```python
corr= data.corr()

plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True)
plt.show()
```

Most of the columns seems pretty correlated except the Reseach Column. Thus dropping the Research column would be a wise decision, as it will reduce some of our training cost.
